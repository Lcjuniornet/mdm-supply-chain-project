## 1Ô∏è‚É£ Setup e Importa√ß√µes

```python
# Importa√ß√µes
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
import warnings
warnings.filterwarnings('ignore')

# Configura√ß√µes de visualiza√ß√£o
plt.style.use('seaborn-v0_8-darkgrid')
sns.set_palette('husl')
%matplotlib inline

# Configura√ß√µes pandas
pd.set_option('display.max_columns', None)
pd.set_option('display.float_format', lambda x: f'{x:,.2f}')

print("‚úÖ Bibliotecas importadas com sucesso!")
print(f"Pandas: {pd.__version__}")
print(f"NumPy: {np.__version__}")
```

```python
# Carregar dataset
df = pd.read_csv('../data/raw/materiais_raw.csv')

print(f"üìä Dataset carregado com sucesso!")
print(f"   Registros: {len(df):,}")
print(f"   Colunas: {len(df.columns)}")
print(f"   Mem√≥ria: {df.memory_usage(deep=True).sum() / 1024**2:.2f} MB")
```

## 3Ô∏è‚É£ Estat√≠sticas Descritivas

```python
# Estat√≠sticas descritivas - campos num√©ricos
print("üìà Estat√≠sticas Descritivas (Campos Num√©ricos):\n")
df.describe()
```

```python
# Estat√≠sticas descritivas - campos categ√≥ricos
print("üìä Estat√≠sticas Descritivas (Campos Categ√≥ricos):\n")
df.describe(include='object')
```

## 4Ô∏è‚É£ An√°lise de Valores em Estoque

```python
# Calcular valor em estoque
df['valor_estoque'] = df['preco_unitario'] * df['estoque_atual']

# Estat√≠sticas valor estoque
valor_total = df['valor_estoque'].sum()
valor_medio = df['valor_estoque'].mean()
valor_mediano = df['valor_estoque'].median()

print(f"üí∞ AN√ÅLISE DE VALOR EM ESTOQUE")
print(f"{'‚îÄ'*60}")
print(f"Valor Total em Estoque: R$ {valor_total:,.2f}")
print(f"Valor M√©dio por Material: R$ {valor_medio:,.2f}")
print(f"Valor Mediano: R$ {valor_mediano:,.2f}")
print(f"\nObserva√ß√£o: M√©dia >> Mediana = Concentra√ß√£o em poucos itens")
```

```python
# Top 10 materiais por valor em estoque
print("üèÜ TOP 10 MATERIAIS POR VALOR EM ESTOQUE:\n")
top10_valor = df.nlargest(10, 'valor_estoque')[[
    'codigo_material', 'descricao', 'categoria', 
    'preco_unitario', 'estoque_atual', 'valor_estoque'
]]
display(top10_valor)
```

```python
# Percentual Top 10 representa do total
valor_top10 = top10_valor['valor_estoque'].sum()
percentual_top10 = (valor_top10 / valor_total) * 100

print(f"\nüìä CONCENTRA√á√ÉO DE VALOR:")
print(f"{'‚îÄ'*60}")
print(f"Top 10 materiais (0,3%): R$ {valor_top10:,.2f}")
print(f"Representa: {percentual_top10:.2f}% do valor total")
print(f"\nüí° INSIGHT: Menos de 1% dos materiais concentram ~5% do valor!")
```

## 5Ô∏è‚É£ An√°lise de Distribui√ß√µes

```python
# Distribui√ß√£o de pre√ßos
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Histograma
axes[0].hist(df['preco_unitario'], bins=50, color='steelblue', alpha=0.7, edgecolor='black')
axes[0].axvline(df['preco_unitario'].mean(), color='red', linestyle='--', 
                linewidth=2, label=f'M√©dia: R$ {df["preco_unitario"].mean():.2f}')
axes[0].axvline(df['preco_unitario'].median(), color='green', linestyle='--', 
                linewidth=2, label=f'Mediana: R$ {df["preco_unitario"].median():.2f}')
axes[0].set_xlabel('Pre√ßo Unit√°rio (R$)', fontweight='bold')
axes[0].set_ylabel('Frequ√™ncia', fontweight='bold')
axes[0].set_title('Distribui√ß√£o de Pre√ßos Unit√°rios', fontweight='bold', fontsize=12)
axes[0].legend()
axes[0].grid(alpha=0.3)

# Boxplot
axes[1].boxplot(df['preco_unitario'], vert=True)
axes[1].set_ylabel('Pre√ßo Unit√°rio (R$)', fontweight='bold')
axes[1].set_title('Boxplot - Identifica√ß√£o de Outliers', fontweight='bold', fontsize=12)
axes[1].grid(alpha=0.3)

plt.tight_layout()
plt.show()

print("\nüí° INSIGHT: Distribui√ß√£o altamente assim√©trica (skewed right)")
print("   Maioria dos materiais tem pre√ßo baixo (<R$ 100)")
print("   Poucos outliers com pre√ßos muito altos (>R$ 1.000)")
```

```python
# Distribui√ß√£o de estoque
fig, ax = plt.subplots(figsize=(12, 5))

ax.hist(df['estoque_atual'], bins=50, color='green', alpha=0.7, edgecolor='black')
ax.axvline(df['estoque_atual'].mean(), color='red', linestyle='--', 
           linewidth=2, label=f'M√©dia: {df["estoque_atual"].mean():.0f} unidades')
ax.axvline(df['estoque_atual'].median(), color='orange', linestyle='--', 
           linewidth=2, label=f'Mediana: {df["estoque_atual"].median():.0f} unidades')
ax.set_xlabel('Estoque Atual (unidades)', fontweight='bold')
ax.set_ylabel('Frequ√™ncia', fontweight='bold')
ax.set_title('Distribui√ß√£o de Estoque Atual', fontweight='bold', fontsize=14)
ax.legend()
ax.grid(alpha=0.3)

plt.tight_layout()
plt.show()
```

## 6Ô∏è‚É£ An√°lise por Categoria

```python
# Estat√≠sticas por categoria
print("üìä ESTAT√çSTICAS POR CATEGORIA:\n")

cat_stats = df.groupby('categoria').agg({
    'codigo_material': 'count',
    'preco_unitario': ['mean', 'median'],
    'estoque_atual': 'sum',
    'valor_estoque': 'sum'
}).round(2)

cat_stats.columns = ['Qtd_Materiais', 'Preco_Medio', 'Preco_Mediano', 
                      'Estoque_Total', 'Valor_Total']
cat_stats = cat_stats.sort_values('Valor_Total', ascending=False)

display(cat_stats)
```

```python
# Visualiza√ß√£o: Valor por Categoria
fig, ax = plt.subplots(figsize=(12, 6))

cat_stats_plot = cat_stats.sort_values('Valor_Total')
bars = ax.barh(cat_stats_plot.index, cat_stats_plot['Valor_Total']/1e6, 
               color='steelblue', alpha=0.8, edgecolor='black')

# Destacar top 3
for i in range(len(bars)-3, len(bars)):
    bars[i].set_color('#e74c3c')
    bars[i].set_alpha(0.9)

ax.set_xlabel('Valor Total em Estoque (R$ Milh√µes)', fontweight='bold', fontsize=11)
ax.set_title('Valor em Estoque por Categoria\n(Vermelho = Top 3)', 
             fontweight='bold', fontsize=14)
ax.grid(axis='x', alpha=0.3)

# Adicionar valores
for i, (bar, val) in enumerate(zip(bars, cat_stats_plot['Valor_Total']/1e6)):
    ax.text(val + 5, bar.get_y() + bar.get_height()/2, 
            f'R$ {val:.1f}M', va='center', fontweight='bold', fontsize=9)

plt.tight_layout()
plt.show()

print(f"\nüí° INSIGHT: Top 3 categorias concentram {(cat_stats_plot.tail(3)['Valor_Total'].sum()/valor_total*100):.1f}% do valor total")
```

## 7Ô∏è‚É£ An√°lise de Correla√ß√µes

```python
# Matriz de correla√ß√£o - campos num√©ricos
campos_numericos = ['preco_unitario', 'estoque_atual', 'estoque_minimo', 'valor_estoque']
correlation_matrix = df[campos_numericos].corr()

print("üîó MATRIZ DE CORRELA√á√ÉO:\n")
display(correlation_matrix)

# Heatmap
fig, ax = plt.subplots(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, fmt='.3f', cmap='coolwarm', 
            center=0, square=True, linewidths=1, cbar_kws={'label': 'Correla√ß√£o'},
            ax=ax)
ax.set_title('Matriz de Correla√ß√£o - Vari√°veis Num√©ricas', 
             fontweight='bold', fontsize=14, pad=15)
plt.tight_layout()
plt.show()

print("\nüí° INSIGHTS CORRELA√á√ïES:")
print("   ‚Ä¢ Pre√ßo √ó Estoque: Correla√ß√£o fraca (independentes)")
print("   ‚Ä¢ Estoque Atual √ó M√≠nimo: Correla√ß√£o positiva (esperado)")
print("   ‚Ä¢ Valor Estoque: Correla√ß√£o forte com Estoque Atual (√≥bvio)")
```

## 8Ô∏è‚É£ Identifica√ß√£o de Outliers

```python
# Outliers de pre√ßo (Z-score method)
from scipy import stats

z_scores = np.abs(stats.zscore(df['preco_unitario']))
outliers_preco = df[z_scores > 3]

print(f"üîç OUTLIERS IDENTIFICADOS (Z-score > 3):\n")
print(f"Total de outliers em pre√ßo: {len(outliers_preco)} ({len(outliers_preco)/len(df)*100:.2f}%)\n")

if len(outliers_preco) > 0:
    print("Top 10 outliers de pre√ßo:\n")
    display(outliers_preco.nlargest(10, 'preco_unitario')[[
        'codigo_material', 'descricao', 'categoria', 'preco_unitario'
    ]])
    
    print(f"\nüí° INSIGHT: {len(outliers_preco)} materiais com pre√ßos anormalmente altos")
    print(f"   Recomenda√ß√£o: Validar se pre√ßos est√£o corretos (poss√≠vel erro cadastral)")
```

```python
# Scatter plot: Pre√ßo √ó Estoque (identificar anomalias)
fig, ax = plt.subplots(figsize=(12, 7))

scatter = ax.scatter(df['estoque_atual'], df['preco_unitario'], 
                     c=df['valor_estoque'], cmap='viridis', 
                     alpha=0.6, s=50, edgecolors='black', linewidth=0.5)

ax.set_xlabel('Estoque Atual (unidades)', fontweight='bold', fontsize=11)
ax.set_ylabel('Pre√ßo Unit√°rio (R$)', fontweight='bold', fontsize=11)
ax.set_title('Rela√ß√£o: Pre√ßo √ó Estoque\n(Cor = Valor Total em Estoque)', 
             fontweight='bold', fontsize=14)
ax.grid(alpha=0.3)

# Colorbar
cbar = plt.colorbar(scatter, ax=ax)
cbar.set_label('Valor em Estoque (R$)', fontweight='bold')

plt.tight_layout()
plt.show()

print("\nüí° INSIGHT: Quadrante superior direito = Materiais cr√≠ticos (alto pre√ßo + alto estoque)")
```

## 9Ô∏è‚É£ An√°lise de Materiais Parados

```python
# Converter datas
df['ultima_movimentacao'] = pd.to_datetime(df['ultima_movimentacao'])
data_referencia = pd.to_datetime('2024-02-25')

# Calcular dias sem movimento
df['dias_sem_movimento'] = (data_referencia - df['ultima_movimentacao']).dt.days

# Materiais parados (>365 dias)
materiais_parados = df[df['dias_sem_movimento'] > 365]
valor_parado = materiais_parados['valor_estoque'].sum()

print(f"‚è∏Ô∏è AN√ÅLISE DE MATERIAIS PARADOS:\n")
print(f"{'‚îÄ'*60}")
print(f"Materiais sem movimento >365 dias: {len(materiais_parados):,} ({len(materiais_parados)/len(df)*100:.2f}%)")
print(f"Valor capital parado: R$ {valor_parado:,.2f}")
print(f"\nüí∞ OPORTUNIDADE: R$ {valor_parado * 0.10:,.2f}/ano (custo capital 10%)")
```

```python
# Top 20 materiais parados por valor
print("\nüèÜ TOP 20 MATERIAIS PARADOS (Maior Valor):\n")
top20_parados = materiais_parados.nlargest(20, 'valor_estoque')[[
    'codigo_material', 'descricao', 'categoria', 
    'dias_sem_movimento', 'valor_estoque'
]]
display(top20_parados)

print(f"\nüí° INSIGHT: Top 20 materiais parados = R$ {top20_parados['valor_estoque'].sum():,.2f}")
print(f"   A√ß√£o: Avaliar obsolesc√™ncia e possibilidade de liquida√ß√£o")
```

## üîü An√°lise Curva ABC

```python
# Classifica√ß√£o ABC
df_sorted = df.sort_values('valor_estoque', ascending=False).copy()
df_sorted['valor_acumulado'] = df_sorted['valor_estoque'].cumsum()
df_sorted['percentual_acumulado'] = (df_sorted['valor_acumulado'] / valor_total) * 100
df_sorted['posicao'] = range(1, len(df_sorted) + 1)
df_sorted['percentual_itens'] = (df_sorted['posicao'] / len(df_sorted)) * 100

# Classificar
def classificar_abc(row):
    if row['percentual_acumulado'] <= 80:
        return 'A'
    elif row['percentual_acumulado'] <= 95:
        return 'B'
    else:
        return 'C'

df_sorted['classe_abc'] = df_sorted.apply(classificar_abc, axis=1)

# Estat√≠sticas ABC
abc_stats = df_sorted.groupby('classe_abc').agg({
    'codigo_material': 'count',
    'valor_estoque': 'sum'
})
abc_stats['percentual_itens'] = (abc_stats['codigo_material'] / len(df)) * 100
abc_stats['percentual_valor'] = (abc_stats['valor_estoque'] / valor_total) * 100
abc_stats.columns = ['Qtd_Materiais', 'Valor_Total', '% Itens', '% Valor']

print("üìä AN√ÅLISE CURVA ABC:\n")
display(abc_stats)

print(f"\nüí° INSIGHT CURVA ABC:")
print(f"   Classe A: {abc_stats.loc['A', '% Itens']:.1f}% dos itens = {abc_stats.loc['A', '% Valor']:.1f}% do valor")
print(f"   Classe B: {abc_stats.loc['B', '% Itens']:.1f}% dos itens = {abc_stats.loc['B', '% Valor']:.1f}% do valor")
print(f"   Classe C: {abc_stats.loc['C', '% Itens']:.1f}% dos itens = {abc_stats.loc['C', '% Valor']:.1f}% do valor")
```

```python
# Gr√°fico Pareto
fig, ax1 = plt.subplots(figsize=(14, 7))

# Barras
color_map = {'A': '#27ae60', 'B': '#f39c12', 'C': '#e74c3c'}
colors = [color_map[classe] for classe in df_sorted['classe_abc']]
ax1.bar(df_sorted['posicao'], df_sorted['valor_estoque']/1e6, 
        color=colors, alpha=0.7, width=1)
ax1.set_xlabel('Materiais (ordenados por valor)', fontweight='bold', fontsize=11)
ax1.set_ylabel('Valor em Estoque (R$ Milh√µes)', fontweight='bold', fontsize=11)

# Linha acumulada
ax2 = ax1.twinx()
ax2.plot(df_sorted['posicao'], df_sorted['percentual_acumulado'], 
         color='darkblue', linewidth=3, label='% Acumulado')
ax2.axhline(y=80, color='green', linestyle='--', linewidth=2, alpha=0.7, label='80% (Classe A)')
ax2.axhline(y=95, color='orange', linestyle='--', linewidth=2, alpha=0.7, label='95% (Classe B)')
ax2.set_ylabel('Percentual Acumulado (%)', fontweight='bold', fontsize=11)
ax2.set_ylim(0, 105)
ax2.legend(loc='lower right')

ax1.set_title('Curva ABC - An√°lise de Pareto\n(Verde=A, Laranja=B, Vermelho=C)', 
              fontweight='bold', fontsize=14, pad=15)
ax1.grid(alpha=0.3)

plt.tight_layout()
plt.show()
```
